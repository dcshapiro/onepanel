# Using horoscopes to manage a stock portfolio

## Loading in libraries

```python
import datetime
import pandas as pd
import requests
import numpy as np
import os
from stocker import Stocker
from bs4 import BeautifulSoup
from urllib.request import urlopen
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from textblob import TextBlob
%matplotlib inline
```

## Scraping horoscope data from the New York Post
Modified from http://www.travisriddle.com/Scraping-Horoscopes/
##### Split into 3 time periods since doing it all at once keeps crashing the droplet.

```python
baseurl = 'http://nypost.com/horoscope/'
signs = ['aries', 'taurus', 'gemini', 'cancer', 'leo', 'virgo', 'libra', 
         'scorpio', 'sagittarius', 'capricorn', 'aquarius', 'pisces']
```

```python
start = pd.datetime(2014, 1, 1)
end = pd.datetime(2015, 12, 31)
time_range = pd.date_range(start, end)
scope = []
day_signs = []
pub_date = []
for day in time_range:
    print (day.strftime('%Y-%m-%d'))
    for sign in signs:
        url = baseurl + sign + '-' + day.strftime('%m-%d-%Y') + '/'
        page = requests.get(url)
        if not page.ok:
            continue
        try:
            content = urlopen(url)
            soup = BeautifulSoup(content,'lxml')
            soup = soup.find('div', 'entry-content')
            soup = soup.find('p').string
            scope.append(soup)
            day_signs.append(sign)
            pub_date.append(day.strftime('%m-%d-%Y'))
        except:
            scope.append(np.nan)   
            day_signs.append(sign)
            pub_date.append(day.strftime('%m-%d-%Y'))
            
    df1 = pd.DataFrame({'horoscope' : scope,
                       'sign' : day_signs,
                       'pub_date' : pub_date})
    df1.to_csv('astroscope1.csv', sep='|')
```

```python
df1 = pd.read_csv('astroscope1.csv', index_col=0, sep='|')
df2 = pd.read_csv('astroscope2.csv', index_col=0, sep='|')
df3 = pd.read_csv('astroscope3.csv', index_col=0, sep='|')

frames = [df1,df2,df3]
df = pd.concat(frames, keys=["horoscope","pub_date","sign"])
```

## Determining whether each horoscope is good or bad

```python
def sentiment(horoscope):
    analysis = TextBlob(str(horoscope).lower())
    if analysis.sentiment.polarity > 0:
        return 1
    elif analysis.sentiment.polarity == 0:
        return 0
    else:
        return -1
```

```python
df["sentiment"] = df.apply(lambda row: sentiment(row["horoscope"]), axis=1)
display(df.describe())
```

```python
print(list(df))
df.to_csv('daily_sentiment_final.csv', sep='|')
df.head(20)
```

## Determining allocations based on the sentiment analysis

```python
sentiments_table = pd.read_csv('daily_sentiment_final.csv', sep='|')
sentiments_table = sentiments_table[sentiments_table.columns[3:]]
```

```python
idx = pd.date_range('01-01-2014', '01-01-2018')
stocks = {'aries':'PG', # Procter & Gamble
          'taurus':'FB', # Facebook
          'gemini':'ADBE', # Adobe
          'cancer':'TSLA', # Tesla
          'leo':'LVS', # Las Vegas Sands
          'virgo':'BBT', # BB&T
          'libra':'CBS', # CBS Corp
          'scorpio':'PEP', # PepsiCo
          'sagittarius':'AIR', # Airbus Group
          'capricorn':'AMZN', # Amazon
          'aquarius':'MSTR', # MicroStrategy
          'pisces':'NWSA'} # News Corp
sentiments_dict = {}
for sign in signs:
    sentiments_dict[sign] = sentiments_table[sentiments_table.sign == sign][['pub_date','sentiment']]
    sentiments_dict[sign].columns = ['pub_date', sign]
sentiments_frame = sentiments_dict['aries']
for sign in signs[1:]:
    sentiments_frame = sentiments_frame.merge(sentiments_dict[sign])
sentiments_frame = sentiments_frame.replace(-1, 0)
sentiments_frame[['pub_date']] = sentiments_frame['pub_date'].apply(lambda x: pd.Timestamp(x))
sentiments_frame.index = sentiments_frame.pub_date
sentiments_frame.drop('pub_date', axis=1, inplace=True)

sentiments_frame = sentiments_frame.reindex(idx, fill_value=0)
sentiments_frame.head()
```

```python
alloc_list = []
for i in range(len(sentiments_frame)):
    day_alloc = []
    if sentiments_frame.iloc[i].values.sum() > 0:
        for j in range(len(signs)):
            day_alloc.append(sentiments_frame.iloc[i,j]/sentiments_frame.iloc[i].values.sum())
    else:
        day_alloc = alloc_list[-1]
    alloc_list.append(day_alloc)
```

```python
alloc_frame = pd.DataFrame(alloc_list)
alloc_frame.index = sentiments_frame.index
alloc_frame.columns = signs
alloc_frame.to_csv('alloc_frame.csv')
display(alloc_frame.head())
```

```python
print('Average allocation for each stock:')
alloc_frame.mean(axis=0)
```

## Loading in each stock's performance

```python
stocker_by_sign = {}
for sign in signs:
    stocker_by_sign[sign] = Stocker(stocks[sign]).stock
    stocker_by_sign[sign][sign] = (stocker_by_sign[sign][['Adj. Open']] - stocker_by_sign[sign][['Adj. Open']].shift(1))/stocker_by_sign[sign][['Adj. Open']].shift(1)
    stocker_by_sign[sign] = stocker_by_sign[sign][['Date', sign]]
    stocker_by_sign[sign][['Date']]= stocker_by_sign[sign]['Date'].apply(lambda x: pd.Timestamp(x))
```

```python
profits_frame = stocker_by_sign['aries'][(stocker_by_sign['aries'].Date>=pd.Timestamp(2014,1,1)) & (stocker_by_sign['aries'].Date<pd.Timestamp(2018,1,1))]
for sign in signs[1:]:
    profits_frame = profits_frame.merge(stocker_by_sign[sign][(stocker_by_sign[sign].Date>=pd.Timestamp(2014,1,1)) & (stocker_by_sign[sign].Date<pd.Timestamp(2018,1,1))])
profits_frame.head()
```

## Calculating returns

```python
## Calculating returns for horoscope bot
percentage_returns = []
for i in range(1, len(profits_frame)):
    weights = np.array(alloc_frame.loc[profits_frame.Date[i-1]])
    true_returns = np.array(profits_frame.iloc[i][1:])
    percentage_returns.append(1 + (weights*true_returns).sum())
money = [100]
for i in range(len(percentage_returns)):
    money.append(money[-1]*percentage_returns[i])
    
cumulative_return = (money[-1] - money[0])/money[0]
days_held = (profits_frame.Date.iloc[-1] - profits_frame.Date.iloc[0]).days
```

```python
print('Psychicbot')
print('Annualized return: ', np.around((1+cumulative_return)**(365.0/days_held) - 1, decimals=3))
print('Final money: ', np.around(money[-1], decimals=2))
```

```python
# Calculating returns if we equally invested in each stock
eqwt_returns = []
for i in range(1, len(profits_frame)):
    weights = np.array([1./12.]*12)
    true_returns = np.array(profits_frame.iloc[i][1:])
    eqwt_returns.append(1 + (weights*true_returns).sum())
money_eqwt = [100]
for i in range(len(eqwt_returns)):
    money_eqwt.append(money_eqwt[-1]*eqwt_returns[i])

cumulative_return_eqwt = (money_eqwt[-1] - money_eqwt[0])/money_eqwt[0]
days_held = (profits_frame.Date.iloc[-1] - profits_frame.Date.iloc[0]).days
```

```python
print('Equal weight')
print('Annualized return: ', np.around((1+cumulative_return_eqwt)**(365.0/days_held) - 1, decimals=3))
print('Final money: ', np.around(money_eqwt[-1], decimals=2))
```

```python
X = money
X2 = money_eqwt
C = profits_frame.Date
plt.figure(figsize=(500,500))
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_xlabel('Date')
ax.set_ylabel('Portfolio Value ($)')



plt.plot(C,X, label='Horoscope bot')
plt.plot(C,X2, label='Equal weight')
plt.legend()
fig.autofmt_xdate()

plt.show()
```
