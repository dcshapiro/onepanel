## Scraping images of CEOs

```python
import pandas as pd
import glob
import time
from subprocess import call


profits_tbl = pd.read_csv('../stock_profits2.csv')
baseKeyword = list(profits_tbl["Symbol"])
keywordList = ["CEO","Chief Executive Officer","CEO profile photo","top executive","profile photo"]

import lemayScraper as ls
for stock in baseKeyword:
    ls.scrapeImages(keywordList, stock)
    #wait 10 seconds
    time.sleep(10)
    #kill all nodejs
    call(["killall", "nodejs"])
    
```
